```python
#importing the libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

###Data preprocessing

```python
movies=pd.read_csv("/content/my_movies.csv")
movies.head()
```

```python
movies.shape
```

```python
col=['V1','V2','V3','V4','V5']
movies=movies.drop(col,axis=1)
```

###Apriori algorithm

```python
frequent_movies = apriori(movies, min_support=0.05, use_colnames=True)
frequent_movies
```

```python
frequent_movies_1 = apriori(movies, min_support=0.1, use_colnames=True)
frequent_movies_1
```

We have got 52 rules for both 5% and 10% support

```python
frequent_movies_2 = apriori(movies, min_support=0.2, use_colnames=True)
frequent_movies_2
```

13 sets of movies accurs more than 20% of times.
*   60% of times Sixth sense is watched and 70% of times the movie Gladiator is watched and so on
*   40% of times the movies Sixth Sense, Gladiator and Patriot are watched.

```python
frequent_movies_3 = apriori(movies, min_support=0.3, use_colnames=True)
frequent_movies_3
```

We have got very few rules for support 0.3 and 0.2. Therefore, I will select 0.1 as min support value for enough rules and better results.


I chose the 0.5 as minimum confidence value. In other words, when movie X is watched, we can say that the probability of watching movie(s) Y is 50% or more.

```python
rules = association_rules(frequent_movies_1, metric="confidence", min_threshold=0.5)
rules
```

Let us filter the data having the movie sets whose lift value is more than 2.

```python
rules=rules[rules['lift']>2]
#Sort the values
rules=rules.sort_values('lift',ascending=False)
rules
```

If we examine our 130th index value,
*   The probability of watching Sixth Sense and LOTR1 is 10%, while the probability of watching LOTR2 and Harry Potter is 10%, and the support for both of them is measured as 10%
*   100% of those who watched Sixth Sense and LOTR1 also watched LOTR2 and Harry Potter1.
*   Their correlation is infinity. Means, we can say these movies are strongly dependent.


```python
rules=rules[['antecedents','consequents','support','confidence','lift']]
rules
```
